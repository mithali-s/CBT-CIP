```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from plotly.subplots import make_subplots
from datetime import datetime
```

```python
import os

current_directory = os.getcwd()
print(f"Current working directory: {current_directory}")
```

```python
 #Get the current working directory
current_directory = os.getcwd()

# Specify the path to the file within the current working directory
file_path = os.path.join(current_directory, "Corona Vaccination Side Effects.xlsx")

# Read the Excel file using pandas
covid_df = pd.read_excel("/content/Corona Vaccination Side Effects.xlsx") # Assuming it's an xlsx file, use pd.read_excel
```

```python
covid_df.columns = covid_df.columns.str.strip()
```

```python
covid_df.info()
```

```python
# Display basic info and first few rows
print("Dataset Information:")
covid_df.info()
print("\nFirst 5 rows:\n", covid_df.head())
```

```python
# Summary statistics
print("\nDescriptive Statistics:")
print(covid_df.describe().T)
```

```python
# Checking for missing values
print("\nMissing Values:")
print(covid_df.isnull().sum())
```

```python
# Correlation heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(covid_df.iloc[:, 8:].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix of Numeric Features")
plt.show()
```

```python
# Standardization of numeric columns
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(covid_df.iloc[:, 8:]), columns=covid_df.iloc[:, 8:].columns)
print("\nStandardized Data Preview:")
print(df_scaled.head())
```

```python
# Feature engineering example: Creating a total side effects score
covid_df['Total Side Effects'] = covid_df.iloc[:, 8:20].sum(axis=1)
print("\nFeature Engineering: Created 'Total Side Effects' column")
print(covid_df[['Total Side Effects']].head())
```

```python
# Boxplot of numeric responses
plt.figure(figsize=(12, 6))
sns.boxplot(data=covid_df.iloc[:, 8:], orient="h", palette="coolwarm")
plt.title("Distribution of Survey Responses")
plt.xlabel("Response Scale (1-5)")
plt.show()
```

```python
# Distribution plots for numerical columns
num_cols = covid_df.iloc[:, 8:].columns
for col in num_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(covid_df[col], bins=20, kde=True, color='blue')
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```
