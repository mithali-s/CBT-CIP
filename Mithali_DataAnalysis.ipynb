```python
vaccine_df=pd.read_csv("/content/covid_vaccine_statewise.csv")
```

```python
covid_df.drop(["Sno","ConfirmedIndianNational","ConfirmedForeignNational","Time"],inplace=True,axis=1)
```

```python
covid_df['Date'] = pd.to_datetime(covid_df['Date'], format='%d/%m/%y', errors='coerce')#If needed
covid_df = covid_df.dropna(subset=['Date'])
covid_df['Date'] = covid_df['Date'].dt.strftime('%Y-%m-%d')
```

```python
covid_df.head()
```

```python
covid_df.info()
```

```python
covid_df['active cases'] = covid_df['Confirmed']-covid_df['Cured']+covid_df['Deaths']
```

```python
covid_df.head()
```

```python
covid_df.tail()
```

```python
statewise = pd.pivot_table(covid_df,values=["Confirmed" , "Cured","Deaths"],
                           index="State/UnionTerritory",aggfunc=max)
```

```python
statewise["recovery rate"]=statewise["Cured"]*100/statewise["Confirmed"]
statewise["mortality rate"]=statewise["Deaths"]*100/statewise["Confirmed"]
```

```python
statewise=statewise.sort_values(by="Confirmed",ascending=False)
```

```python
statewise.style.background_gradient(cmap="cubehelix")
```

```python
top_10_active_cases = covid_df.groupby(by="State/UnionTerritory").max()[["active cases", "Date"]].sort_values(
by=["active cases"], ascending=False).reset_index()
top_10_active_cases = top_10_active_cases.iloc[:10]
fig = plt.figure(figsize=(16, 9))
plt.title("Top 10 States with Active COVID-19 Cases in India")
ax = sns.barplot(data=top_10_active_cases, y="active cases", x="State/UnionTerritory", linewidth=2, edgecolor="red")
plt.xlabel("States/Union Territories")
plt.ylabel("Active Cases")
plt.show()
```

```python
top_10_death_cases = covid_df.groupby(by="State/UnionTerritory").max()[["Deaths", "Date"]].sort_values(
    by=["Deaths"], ascending=False).reset_index()
top_10_death_cases = top_10_death_cases.iloc[:12]
fig = plt.figure(figsize=(16, 9))
plt.title("Top 10 States with COVID-19 Death Cases in India")
ax = sns.barplot(data=top_10_death_cases, y="Deaths", x="State/UnionTerritory", linewidth=2, edgecolor="red")
plt.xlabel("States/Union Territories")
plt.ylabel("Death Cases")
plt.show()
```

```python
fig = plt.figure(figsize=(12, 6))


ax = sns.lineplot(data=covid_df[covid_df["State/UnionTerritory"].isin(["Maharashtra", "Delhi", "Gujarat", "Tamil Nadu", "West Bengal"])],
                  x="Date", y="active cases", hue="State/UnionTerritory")

ax.set_title("Top 5 Affected States in India", size=16)
plt.show()
```

```python
vaccine=vaccine_df[vaccine_df.State!="India"]
vaccine
```

```python
#most vaccinated state
max_vac = vaccine.groupby("State")["Total Individuals Vaccinated"].sum().to_frame("Total Individuals Vaccinated")
max_vac = max_vac.sort_values("Total Individuals Vaccinated",ascending=False)[:5]
max_vac
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig=plt.figure(figsize=(10,5))
plt.title("Top 5 vaccinated states in India",size=20)
x=sns.barplot(data=max_vac.iloc[:10], y = max_vac["Total Individuals Vaccinated"],x=max_vac.index,linewidth=2,edgecolor="blue")
plt.ylabel("vaccination")
plt.xlabel("States")
plt.show()
plt.show()
```

```python
min_vac = vaccine.groupby("State")["Total Individuals Vaccinated"].sum().to_frame("Total Individuals Vaccinated")
min_vac = min_vac.sort_values("Total Individuals Vaccinated",ascending=True)[:5]
min_vac
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig=plt.figure(figsize=(10,5))
plt.title("Least 5 vaccinated states in India",size=20)
x=sns.barplot(data=min_vac.iloc[:10], y = min_vac["Total Individuals Vaccinated"],x=min_vac.index,linewidth=2,edgecolor="blue")
plt.ylabel("vaccination")
plt.xlabel("States")
plt.show()
```
