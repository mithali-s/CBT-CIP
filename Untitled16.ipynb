```python
cols = ['ConfirmedIndianNational', 'ConfirmedForeignNational']
for col in cols:
    covid_df[col] = covid_df[col].replace('-', '0').pipe(pd.to_numeric, errors='coerce')

print('Head:\n', covid_df.head(), '\nInfo:\n', covid_df.info(), '\nComplete')
```

```python
numeric_cols = ['ConfirmedIndianNational', 'ConfirmedForeignNational', 'Cured', 'Deaths', 'Confirmed']
correlation_matrix = covid_df[numeric_cols].corr()
print('Correlation Matrix:')
print(correlation_matrix)
# Plot the correlation matrix using seaborn heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
X = covid_df[numeric_cols]
vif_data = pd.DataFrame()
vif_data["Feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
print('\nVariance Inflation Factor (VIF) for Multicollinearity:')
print(vif_data)
```

```python
from sklearn.preprocessing import StandardScaler

# Select the numeric columns to standardize
numeric_cols = ['ConfirmedIndianNational', 'ConfirmedForeignNational', 'Cured', 'Deaths', 'Confirmed']

# Standardize the numeric columns
covid_df[numeric_cols] = StandardScaler().fit_transform(covid_df[numeric_cols])

# Show the head of the updated dataframe
print('Head of Standardized DataFrame:')
print(covid_df.head())

print('Data Standardization complete.')
```

```python
# New feature: Active Cases
covid_df['Active'] = covid_df['Confirmed'] - (covid_df['Deaths'] + covid_df['Cured'])

# New feature: Ratios (avoiding division by zero)
covid_df['Death_Ratio'] = np.where(covid_df['Confirmed'] != 0, covid_df['Deaths'] / covid_df['Confirmed'], 0)
covid_df['Cured_Ratio'] = np.where(covid_df['Confirmed'] != 0, covid_df['Cured'] / covid_df['Confirmed'], 0)

# Show the head of the updated dataframe
print('Engineered Features:\n',
      covid_df[['Sno', 'Date', 'State/UnionTerritory', 'Confirmed',
          'Cured', 'Deaths', 'Active', 'Death_Ratio', 'Cured_Ratio']].head())

print('Feature engineering complete.')
```

```python

covid_df['active cases'] = covid_df['Confirmed']-covid_df['Cured']+covid_df['Deaths']
covid_df.head()
covid_df.tail()
```

```python
statewise = pd.pivot_table(covid_df,values=["Confirmed" , "Cured","Deaths"],
                           index="State/UnionTerritory",aggfunc=max)
statewise["recovery rate"]=statewise["Cured"]*100/statewise["Confirmed"]
statewise["mortality rate"]=statewise["Deaths"]*100/statewise["Confirmed"]
statewise=statewise.sort_values(by="Confirmed",ascending=False)
statewise.style.background_gradient(cmap="cubehelix")
```

```python
top_10_active_cases = covid_df.groupby(by="State/UnionTerritory").max()[["active cases", "Date"]].sort_values(
by=["active cases"], ascending=False).reset_index()
top_10_active_cases = top_10_active_cases.iloc[:10]
fig = plt.figure(figsize=(16, 9))
plt.title("Top 10 States with Active COVID-19 Cases in India")
ax = sns.barplot(data=top_10_active_cases, y="active cases", x="State/UnionTerritory", linewidth=2, edgecolor="red")
plt.xlabel("States/Union Territories")
plt.ylabel("Active Cases")
plt.show()
```

```python
top_10_death_cases = covid_df.groupby(by="State/UnionTerritory").max()[["Deaths", "Date"]].sort_values(
    by=["Deaths"], ascending=False).reset_index()
top_10_death_cases = top_10_death_cases.iloc[:12]
fig = plt.figure(figsize=(16, 9))
plt.title("Top 10 States with COVID-19 Death Cases in India")
ax = sns.barplot(data=top_10_death_cases, y="Deaths", x="State/UnionTerritory", linewidth=2, edgecolor="red")
plt.xlabel("States/Union Territories")
plt.ylabel("Death Cases")
plt.show()
```
